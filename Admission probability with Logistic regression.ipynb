```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
raw=pd.read_csv('E:/Notes/machine-learning-ex2/ex2data1.csv')
raw.head()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='exam1 score',y='exam2 score',data=raw,hue='status')
plt.style.use('ggplot')
```

```python
data=np.array(raw)
X=data[:,:2]
y=data[:,-1]
```

```python
lr=LogisticRegression()
lr.fit(X,y)
```

```python
lr.predict(np.array([45,85]).reshape(1,-1))
```

```python
lr.predict_proba(np.array([45,85]).reshape(1,-1))
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='exam1 score',y='exam2 score',data=raw,hue='status')
plt.style.use('ggplot')
w = lr.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(20, 100,100)
yy = a * xx - (lr.intercept_[0]) / w[1]

plt.plot(xx, yy, 'k-')
```
